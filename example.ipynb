## OpenPIV Correlation

```python
from openpiv import windef, tools, scaling, validation, filters, preprocess, settings
from defdap import hrdic
import numpy as np
from defdap import hrdic

import matplotlib.pyplot as plt
%matplotlib inline
```

## Define images and settings

```python
mysettings = settings.PIVSettings()

mysettings.filepath_images = '../TANIST_process/Output/Test/'

mysettings.frame_pattern_a = 'Zone_BSE_Step_0.tif'

mysettings.frame_pattern_b = 'Zone_BSE_Step_4.tif'

mysettings.save_directory = '../TANIST_process/Output/'

mysettings.save_filename = 'Test_1.txt'

mysettings.max_array_size = 10000000

mysettings.windowsizes = (1024, 512, 256, 128)

mysettings.num_iterations = len(mysettings.windowsizes)

mysettings.overlap = (0, 0, 0, 0)
```

## Run correlation

```python
# run the script with the given settings
windef.piv(mysettings)
```

## Display result

```python
path = str(settings.save_directory) + '/' + settings.save_filename

dic_map = hrdic.Map(path, data_type = 'openpiv')

#shape = 10656. 10912.
#binning = 64
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10,8))
dic_map.plot_map('max_shear', vmax = 0.2, fig = fig, ax = ax)
```
