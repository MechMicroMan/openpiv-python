## OpenPIV Correlation

```python
from openpiv import windef, tools, scaling, validation, filters, preprocess, settings
from defdap import hrdic
import numpy as np
from defdap import hrdic

import matplotlib.pyplot as plt
%matplotlib inline
```

## Define images and settings

```python
mysettings = settings.PIVSettings()

mysettings.filepath_images = '//10.99.237.2/Tescan_users/Rhys/'
mysettings.frame_pattern_a = 'Lucy_Ti64_uncharged_003.tif'
mysettings.frame_pattern_b = 'Lucy_Ti64_uncharged_004.tif'

mysettings.save_directory = '../TANIST_process/Output/'
mysettings.save_filename = 'Test_2.txt'
mysettings.max_array_size = 10000000
mysettings.windowsizes = (512, 256, 128, 64, 32)

mysettings.num_iterations = len(mysettings.windowsizes)
mysettings.overlap = (0, 0, 0, 0, 0)
```

## Run correlation

```python
# run the script with the given settings
windef.piv(mysettings)
```

## Display result

```python
path = str(mysettings.save_directory) + '/' + mysettings.save_filename

dic_map = hrdic.Map(path, data_type = 'openpiv')

#shape = 10656. 10912.
#binning = 64
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10,8))
dic_map.plot_map('e', (1,1), vmin=-0.01, vmax=0.01, cmap='coolwarm', fig = fig, ax = ax)
plt.show()
```
